```python
# Import packages
from xmitgcm import open_mdsdataset
import pylab as plt
```

```python
### Import data
# Basic tests
data_dyn_lsr=open_mdsdataset('./tr_run.dyn_lsr/',prefix=['UICE','VICE'],delta_t=1800)
data_dyn_jfnk=open_mdsdataset('./tr_run.dyn_jfnk/',prefix=['UICE','VICE'],delta_t=1800)

# Elliptical yield curve non-normal flow rule
data_dyn_lsr_ellnnfr=open_mdsdataset('./tr_run.dyn_lsr_ellnnfr/',prefix=['UICE','VICE'],delta_t=1800)
# data.dyn_jfnk_ellnnfr=open_mdsdataset('./tr_run.dyn_jfnk_ellnnfr/',prefix=['UICE','VICE'],delta_t=1800)

# Teardrop yield curve
data_dyn_jfnk_teardrop=open_mdsdataset('./tr_run.dyn_jfnk_teardrop/',prefix=['UICE','VICE'],delta_t=1800)
data_dyn_lsr_teardrop=open_mdsdataset('./tr_run.dyn_lsr_teardrop/',prefix=['UICE','VICE'],delta_t=1800)

# Mohr-Coulomb familly of yield curves
data_dyn_lsr_fmc=open_mdsdataset('./tr_run.dyn_lsr_fmc/',prefix=['UICE','VICE'],delta_t=1800)
data_dyn_lsr_mce=open_mdsdataset('./tr_run.dyn_lsr_mce/',prefix=['UICE','VICE'],delta_t=1800)
```

```python
# plot reference U & V fields

v_max=data_dyn_jfnk['UICE'][1].values.max()

fig,ax=plt.subplots()
data_dyn_jfnk['UICE'][1].plot(vmax=v_max,vmin=-v_max,cmap='RdBu_r')
ax.set_aspect('equal')
ax.set_title('UICE JNFK')

fig,ax=plt.subplots()
data_dyn_jfnk['VICE'][1].plot(cmap='RdBu_r')
ax.set_aspect('equal')
ax.set_title('VICE JNFK')
```

```python
## Definition differences UICE
duice_lsr_jfnk=data_dyn_lsr['UICE']-data_dyn_jfnk['UICE']
duice_lsr_ell_ellnnfr=data_dyn_lsr_ellnnfr['UICE']-data_dyn_lsr['UICE']

duice_lsr_ell_teardrop=data_dyn_lsr_teardrop['UICE']-data_dyn_lsr['UICE']
duice_teardrop_lsr_jfnk=data_dyn_lsr_teardrop['UICE']-data_dyn_jfnk_teardrop['UICE']
duice_jfnk_ell_teardrop=data_dyn_jfnk['UICE']-data_dyn_jfnk_teardrop['UICE']

duice_lsr_ell_fmc=data_dyn_lsr_fmc['UICE']-data_dyn_lsr['UICE']
duice_lsr_ell_mce=data_dyn_lsr_mce['UICE']-data_dyn_lsr['UICE']

## Definition differences VICE
dvice_lsr_jfnk=data_dyn_lsr['VICE']-data_dyn_jfnk['VICE']
dvice_lsr_ell_ellnnfr=data_dyn_lsr_ellnnfr['VICE']-data_dyn_lsr['VICE']

dvice_lsr_ell_teardrop=data_dyn_lsr_teardrop['VICE']-data_dyn_lsr['VICE']
dvice_teardrop_lsr_jfnk=data_dyn_lsr_teardrop['VICE']-data_dyn_jfnk_teardrop['VICE']
dvice_jfnk_ell_teardrop=data_dyn_jfnk['VICE']-data_dyn_jfnk_teardrop['VICE']

dvice_lsr_ell_fmc=data_dyn_lsr_fmc['VICE']-data_dyn_lsr['VICE']
dvice_lsr_ell_mce=data_dyn_lsr_mce['VICE']-data_dyn_lsr['VICE']
```

```python
## Plotting diff uice
v_max=0.1

fig,ax=plt.subplots()
duice_lsr_jfnk[1].plot(vmax=v_max,vmin=-v_max,cmap='RdBu_r')
ax.set_aspect('equal')
ax.set_title('diff UICE ELL: LSR JNFK')

fig,ax=plt.subplots()
duice_lsr_ell_ellnnfr[1].plot(vmax=v_max,vmin=-v_max,cmap='RdBu_r')
ax.set_aspect('equal')
ax.set_title('diff UICE LSR: ELL ELLNNFR')

fig,ax=plt.subplots()
duice_lsr_ell_teardrop[1].plot(vmax=v_max,vmin=-v_max,cmap='RdBu_r')
ax.set_aspect('equal')
ax.set_title('diff UICE LSR: ELL TEARDROP')

fig,ax=plt.subplots()
duice_teardrop_lsr_jfnk[1].plot(vmax=v_max,vmin=-v_max,cmap='RdBu_r')
ax.set_aspect('equal')
ax.set_title('diff UICE TEARDROP: LSR JNFK')

fig,ax=plt.subplots()
duice_jfnk_ell_teardrop[1].plot(vmax=v_max,vmin=-v_max,cmap='RdBu_r')
ax.set_aspect('equal')
ax.set_title('diff UICE JFNK: ELL TEARDROP')

fig,ax=plt.subplots()
duice_lsr_ell_fmc[1].plot(vmax=v_max,vmin=-v_max,cmap='RdBu_r')
ax.set_aspect('equal')
ax.set_title('diff UICE LSR: ELL FMC')

fig,ax=plt.subplots()
duice_lsr_ell_mce[1].plot(vmax=v_max,vmin=-v_max,cmap='RdBu_r')
ax.set_aspect('equal')
ax.set_title('diff UICE LSR: ELL MCE')
```

```python
## Plotting diff vice
v_max=0.1

fig,ax=plt.subplots()
dvice_lsr_jfnk[1].plot(vmax=v_max,vmin=-v_max,cmap='RdBu_r')
ax.set_aspect('equal')
ax.set_title('diff VICE ELL: LSR JNFK')

fig,ax=plt.subplots()
dvice_lsr_ell_ellnnfr[1].plot(vmax=v_max,vmin=-v_max,cmap='RdBu_r')
ax.set_aspect('equal')
ax.set_title('diff VICE LSR: ELL ELLNNFR')

fig,ax=plt.subplots()
dvice_lsr_ell_teardrop[1].plot(vmax=v_max,vmin=-v_max,cmap='RdBu_r')
ax.set_aspect('equal')
ax.set_title('diff VICE LSR: ELL TEARDROP')

fig,ax=plt.subplots()
dvice_teardrop_lsr_jfnk[1].plot(vmax=v_max,vmin=-v_max,cmap='RdBu_r')
ax.set_aspect('equal')
ax.set_title('diff VICE TEARDROP: LSR JNFK')

fig,ax=plt.subplots()
dvice_jfnk_ell_teardrop[1].plot(vmax=v_max,vmin=-v_max,cmap='RdBu_r')
ax.set_aspect('equal')
ax.set_title('diff VICE JFNK: ELL TEARDROP')

fig,ax=plt.subplots()
dvice_lsr_ell_fmc[1].plot(vmax=v_max,vmin=-v_max,cmap='RdBu_r')
ax.set_aspect('equal')
ax.set_title('diff VICE LSR: ELL FMC')

fig,ax=plt.subplots()
dvice_lsr_ell_mce[1].plot(vmax=v_max,vmin=-v_max,cmap='RdBu_r')
ax.set_aspect('equal')
ax.set_title('diff VICE LSR: ELL MCE')

```
